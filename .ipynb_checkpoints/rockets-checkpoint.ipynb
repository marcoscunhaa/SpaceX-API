```python
#Aplicado a raspagem de dados dos foguetes da Space X
import requests
import json

url = 'https://api.spacexdata.com/v3/rockets'

response = requests.get(url)

if response.status_code == 200:
    data = json.loads(response.text)
    print(json.dumps(data, indent=4))
else:
    print('Error retrieving data from the SpaceX API.')
```

```python
#Filtrando apenas os dados relevantes
rocket_name = []
cost_per_launch = []
country = []
company = []
height = []
diameter = []
first_flight = []
success_rate_pct = []


for rockets in data:
    rocket_name.append(rockets['rocket_name'])
    cost = float(rockets['cost_per_launch'])
    cost_str = format(cost, ',').replace(',', '.')
    cost_per_launch.append(cost_str)
    country.append(rockets['country'])
    company.append(rockets['company'])
    height.append(rockets['height']['meters'])
    diameter.append(rockets['diameter']['meters'])
    first_flight.append(rockets['first_flight'])
    success_rate_pct.append(rockets['success_rate_pct']/100)

display(rocket_name)
display(cost_per_launch)
display(country)
display(company)
display(height)
display(diameter)
display(first_flight)
display(success_rate_pct)
```

```python
#Implementando um data frame com os dados filtrados
import pandas as pd

data_rocket = {
    'Rocket Name': rocket_name,
    'Cost Per Launch($)': cost_per_launch,
    'Country of Manufacture': country,
    'Company': company,
    'Height(M)': height,
    'Diameter(M)': diameter,
    'First Flight': first_flight,
    'Success Rate Pct(%)': success_rate_pct
}

df = pd.DataFrame(data_rocket)
df['First Flight'] = pd.to_datetime(df['First Flight'])

display(df)
```
